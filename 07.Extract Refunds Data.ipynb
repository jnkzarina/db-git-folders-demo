```sql
CREATE TABLE IF NOT EXISTS gizmobox.bronze.refunds (  
    refund_id INT,  
    payment_id INT NOT NULL,  
    refund_timestamp date NOT NULL,  
    refund_amount DECIMAL(10, 2) NOT NULL,  
    refund_reason STRING NOT NULL,
    CONSTRAINT pk_refund_id PRIMARY KEY (refund_id)
)
USING DELTA;
```
