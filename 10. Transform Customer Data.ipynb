```python
%sql
CREATE TABLE gizmobox.silver.customers
AS
with cte as (
  SELECT
    CUSTOMER_ID,
    MAX(created_timestamp) max_created_timestamp
  FROM
    gizmobox.bronze.v_customers
  where
    customer_id is not null
  GROUP BY
    customer_id
)
select
  distinct 
  cst.customer_id,
  cst.customer_name,
  cast(cst.date_of_birth as date) as date_of_birth,
  cast(cst.created_timestamp as timestamp) as created_timestamp,
  cst.email,
  cast(cst.member_since as date) as member_since,
  cst.telephone
from
  gizmobox.bronze.v_customers cst
    inner join cte
      on cst.customer_id = cte.customer_id
      and cst.created_timestamp = cte.max_created_timestamp;
```
