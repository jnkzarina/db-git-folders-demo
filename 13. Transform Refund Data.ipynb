```python
%sql
Create table if not exists gizmobox.silver.refunds
as
select
  ref.refund_id,
  ref.payment_id,
  date_format(ref.refund_timestamp,'yyyy-MM-dd') as refund_date,
  date_format(ref.refund_timestamp,'HH:mm:ss') as refund_time,
  ref.refund_amount,
  regexp_extract(ref.refund_reason,'^(.*?):',1) as refund_reason,
  regexp_extract(ref.refund_reason,':(.*?)$',1) as refund_source
from
  gizmobox.bronze.refunds ref;
```
