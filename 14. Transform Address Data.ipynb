```python
%sql
create table if not exists gizmobox.silver.addresses
as 
SELECT * 
FROM (
    SELECT customer_id,
    address_type,
    address_line_1,
    city,
    state,
    postcode
    from gizmobox.bronze.v_addresses
)
PIVOT (MAX(address_line_1) as address_line_1,
max(city) as city,
max(state) as state,
max(postcode) as postcode
for address_type in ('shipping','billing')
);

```
