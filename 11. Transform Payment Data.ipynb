```python
%sql
create table gizmobox.silver.payments
as 
select
  pmt._c0 as payment_id,
  pmt._c1 as order_id,
  date_format(pmt._c2, 'yyyy-MM-dd') as payment_date,
  date_format(pmt._c2, 'HH:mm:ss') as payment_time,
  case pmt._c3
    when 1 then 'Success'
    when 2 then 'Pending'
    when 3 then 'Cancelled'
    when 4 then 'Failed'
  end as payment_status,
  pmt._c4 as payment_method
from
  gizmobox.bronze.v_payments pmt;
```

```python
%sql
select * from gizmobox.silver.payments;
```
